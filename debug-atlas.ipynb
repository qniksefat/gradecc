```python
data_rs = load_data(subj=1, epic='rest')
# just took a subj to make atlas
surf_labels, mask_removed = load_atlas(data_rs)
# make average corr_mat
```

```python
text_bar, color_map, data_to_show, data_range = \
        'schafer', 'tab20', surf_labels, None

p = Plot(surf_lh=surf_lh, surf_rh=surf_rh,
        size=(1600, 300),
        layout='row',
        label_text=[text_bar],
        )

p.add_layer(
data_to_show, cbar=True, cmap=color_map, color_range=data_range,
)

fig = p.build()
```
